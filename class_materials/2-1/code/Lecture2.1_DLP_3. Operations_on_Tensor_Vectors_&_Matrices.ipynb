# **1. Tensor의 Norm**

```python
import torch
```

**1.2 Ln Norm**

```python
# L1 Norm

a = torch.tensor([4, 3], dtype=torch.float32)

l1_norm = torch.norm(a, p=1)

print(f'L1 Norm of a: {l1_norm}')
```

```python
# L2 Norm
a = torch.tensor([4, 3], dtype=torch.float32)

l2_norm = torch.norm(a, p=2)

print(f'L2 Norm of a: {l2_norm}')
```

```python
# L∞ Norm
a = torch.tensor([4, 3], dtype=torch.float32)

linf_norm = torch.norm(a, p=float('inf'))

print(f'L∞ Norm of a: {linf_norm}')
```

```python
a = torch.tensor([4, 3], dtype=torch.float32)

linf_norm = torch.max(a.abs())

print(f'L∞ Norm of a: {linf_norm}')
```

# **2. 유사도**

**2.2 맨해튼 유사도**

```python
# 맨해튼 유사도의 코드 표현 실습

b = torch.tensor([1, 0, 2], dtype=torch.float32)
c = torch.tensor([0, 1, 2], dtype=torch.float32)

# 맨해튼 거리 계산
manhattan_distance = torch.norm(b - c, p = 1)

# 유사도 계산 (예시: 거리의 역수)
manhattan_similarity = 1 / (1 + manhattan_distance)

print(f'Manhattan Distance: {manhattan_distance}')
print(f'Manhattan Similarity: {manhattan_similarity}')
```

**2.3 유클리드 유사도**

```python
# 유클리드 유사도의 코드 표현 실습

# 유클리드 거리 계산
euclidean_distance = torch.norm(b - c, p = 2)

# 유사도 계산 (예시: 거리의 역수)
euclidean_similarity = 1 / (1 + euclidean_distance)

print(f'Euclidean Distance: {euclidean_distance.item()}')
print(f'Euclidean Similarity: {euclidean_similarity.item()}')
```

**2.4 코사인 유사도**

```python
# 코사인 유사도의 코드 표현 실습

# 코사인 유사도 계산
dot_product = torch.dot(b, c)
norm_b = torch.norm(b, p = 2)
norm_c = torch.norm(c, p = 2)
cosine_similarity = dot_product / (norm_b * norm_c)

print(f'Cosine Similarity: {cosine_similarity.item()}')
```

```python
dot_product = torch.dot(b, c)
cosine_similarity = torch.dot(b, c) / (torch.norm(b, p = 2) * torch.norm(c, p = 2))

print(f'Cosine Similarity: {cosine_similarity.item()}')
```

**3.2 2-D Tensor의 행렬 곱셈 연산**

```python
# 행렬의 곱셈의 코드 표현 실습

D = torch.tensor([[1, 1, 3],
                  [4, 5, 6],
                  [7, 8, 9]])

E = torch.tensor([[1, 0],
                  [1, -1],
                  [2, 1]])

F = D.matmul(E)

print('D.matmul(E) =', F)
```

```python
F = D.mm(E)

print('D.mm(E) =', F)
```

```python
D = torch.tensor([[1, 1, 3],
                  [4, 5, 6],
                  [7, 8, 9]])

E = torch.tensor([[1, 0],
                  [1, -1],
                  [2, 1]])

F = D @ E

print('D @ E =', F)
```

**3.3 2-D Tensor의 행렬 곱셈 연산 활용**

```python
# 흑백 이미지의 대칭 이동
# 흑백 이미지의 좌우로 대칭 이동 코드 표현

G = torch.tensor([[255, 114, 140],
                  [39, 255, 46],
                  [61, 29, 255]])
print('G = ', G)
```

```python
import matplotlib.pyplot as plt

plt.xticks([]), plt.yticks([])
plt.imshow(G, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
H = torch.tensor([[0, 0, 1],
                  [0, 1, 0],
                  [1, 0, 0]])
print('H = ', H)
```

```python
I = G @ H

print('I = ', I)
```

```python
plt.xticks([]), plt.yticks([])
plt.imshow(I, cmap='gray', vmin=0, vmax=255)
plt.show()
```
