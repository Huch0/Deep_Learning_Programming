# **1. Tensor의 모양변경2**

```python
import torch
```

**1.1 cat() 함수를 활용한 Tensor들 간의 연결**

```python
# stack() 함수를 활용한 tensor들 간의 결합에 대한 코드 표현 실습 복습

red_channel = torch.tensor([[255, 0],
                            [0, 255]])
green_channel = torch.tensor([[0, 255],
                              [0, 255]])
blue_channel = torch.tensor([[0, 0],
                             [255, 0]])

a = torch.stack((red_channel, green_channel, blue_channel))

print('a =', a)
print('a.shape =', a.shape)
```

```python
a = torch.stack((red_channel, green_channel, blue_channel), dim = 1)

print('a =', a)
print('a.shape =', a.shape)
```

```python
a = torch.stack((red_channel, green_channel, blue_channel), dim = 2)

print('a =', a)
print('a.shape =', a.shape)
```

```python
# cat() 함수를 활용한 Tensor 간의 연결에 대한 코드 표현 실습 1

b = torch.tensor([[0, 1],
                  [2, 3]])
c = torch.tensor([[4, 5]])

print('b =', b)
print('b.shape =', b.shape)
print('c =', c)
print('c.shape =', c.shape)
```

```python
d = torch.cat((b, c))
print('d =', d)
print('d.shape =', d.shape)
```

```python
# cat() 함수를 활용한 Tensor 간의 연결에 대한 코드 표현 실습 2

e = torch.cat((b, c), 1)
print('e =', e)
```

```python
e = torch.cat((b, c.reshape(2, 1)), 1)
print('e =', e)
print('e.shape =', e.shape)
```

**1.2 expand() 메서드를 활용한 Tensor의 크기 확장**

```python
# expand() 메서드를 활용한 Tensor의 크기를 확장하는 코드 표현 실습

f = torch.tensor([[1, 2, 3]])

print('f =', f)
print('f.shape =', f.shape)
```

```python
g = f.expand(4, 3)

print('g =', g)
print('g.shape =', g.shape)
```

**1.3 repeat() 메서드를 활용한 Tensor의 크기 확장**

```python
# repeat()메서드를 활용한 Tensor의 크기를 확장하는 코드 표현 실습

h = torch. tensor([[1, 2],
                   [3, 4]])

print('h =', h)
print('h.shape =', h.shape)
```

```python
i = h.repeat(2, 3)

print('i =' , i)
print('i.shape =', i.shape)
```

# **2. Tensor의 기초 연산**

**2.1 Tensor의 산술연산**

```python
# 더하기 연산 코드 표현 실습

a = torch.tensor([[1, 2],
                  [3, 4]])
b = torch.tensor([[1, 3],
                  [5, 7]])

print('a+b =', torch.add(a, b))
```

```python
print('id(a) =', id(a))
```

```python
id(b)
print('id(b) =', id(b))
```

```python
id(torch.add(a, b))
print('id(a+b) =', id(torch.add(a, b)))
```

```python
# in-place 방식을 활용한 더하기 연산 코드 표현 실습

print('id(a) =', id(a))
print('a.add_(b) =', a.add_(b))
```

```python
# 연산 후 메모리 상에서 유지하는 위치의 고유한 식별자 확인 코드 표현 실습

id(a.add_(b))
print('id(a.add_(b)) =', id(a.add_(b)))
```

```python
# 크기가 다른 Tensor의 더하기 연산 코드 표현 실습

c = torch.tensor([[1, 2],
                  [4, 5]])
d = torch.tensor([1, 3])

print('c+d =', torch.add(c, d))
```

```python
# 빼기 연산 코드 표현 실습

e = torch.tensor([[2, 3],
                  [4, 5]])
f = torch.tensor([[2, 4],
                  [6, 8]])

print('f-e =', torch.sub(f, e))
```

```python
print('f.sub_(e) =', f.sub_(e))
```

```python
g = torch.tensor([[2, 3],
                  [5, 6]])
h = torch.tensor([[1],
                  [4]])

print('g-h =', torch.sub(g, h))
```

```python
# 스칼라곱 연산 코드 표현 실습

i = 2
j = torch.tensor([[1, 2],
                  [3, 4]])

print('i*j =', torch.mul(i, j))
```

```python
# 요소별 곱하기 연산 코드 표현 실습

k = torch.tensor([[1, 2],
                  [1, 3]])
l = torch.tensor([[2, 3],
                  [1, 4]])

print('k*l =', torch.mul(k, l))
```

```python
print('k.mul_(l) =', k.mul_(l))
```

```python
m = torch.tensor([[1, 4],
                  [7, 8]])
n = torch.tensor([2, 3])

print('m*n =', torch.mul(m, n))
```

```python
# 요소별 나누기 연산 코드 표현 실습

o = torch.tensor([[18, 9],
                  [10, 4]])
p = torch.tensor([[6, 3],
                  [5, 2]])

print('o/p =', torch.div(o, p))
```

```python
print('o.div_(p) =', o.div_(p))
```

```python
o = o.float()
p = p.float()

print('o.div_(p) =', o.div_(p))
```

```python
q = torch.tensor([[12, 8],
                  [9, 4]])
r = torch.tensor([3, 2])

print('q/r =', torch.div(q, r))
```

```python
# 요소별 거듭제곱 연산 코드 표현 실습

s = torch.tensor([[1, 2],
                  [3, 4]])

print('s^2 =', torch.pow(s, 2))
print('s^3 =', torch.pow(s, 3))
```

```python
t = torch.tensor([[5, 4],
                  [3, 2]])
u = torch.tensor([[1, 2],
                  [3, 4]])

print('t**u =', torch.pow(t, u))
```

```python
print('t.pow_(u) = ', t.pow_(u))
```

```python
# 거듭제곱근 연산 코드 표현 실습

s = torch.tensor([[1, 2],
                  [3, 4]])

print('s^(1/2) =', torch.pow(s, 1/2))
print('s^(1/3) =', torch.pow(s, 1/3))
```

**2.2 Tensor의 비교연산**

```python
# 비교연산 코드 표현 실습 1

v = torch.tensor([1, 3, 5, 7])
w = torch.tensor([2, 3, 5, 7])

print('v==w :', torch.eq(v, w))
```

```python
v = torch.tensor([1, 3, 5, 7])
w = torch.tensor([2, 3, 5, 7, 10])

print('v==w :', torch.eq(v, w))
```

```python
# 비교연산 코드 표현 실습 2

print('v!=w :', torch.ne(v, w))
```

```python
print('v>w :' , torch.gt(v, w))
```

```python
print('v>=w :', torch.ge(v, w))
```

```python
print('v<w :', torch.lt(v, w))
```

```python
print('v=<w :', torch.le(v, w))
```

**2.3 Tensor의 논리연산**

```python
# 논리곱(AND) 연산 코드 표현 실습

x = torch.tensor([True, True, False, False])
y = torch.tensor([True, False, True, False])

print('and(x, y) =', torch.logical_and(x, y))
```

```python
# 논리합(OR) 연산 코드 표현 실습

x = torch.tensor([True, True, False, False])
y = torch.tensor([True, False, True, False])

print('or(x, y) =', torch.logical_or(x, y))
```

```python
# 배타적 논리합(XOR) 연산 코드 표현 실습

e = torch.tensor([True, True, False, False])
f = torch.tensor([True, False, True, False])

print('xor(x, y) =', torch.logical_xor(x, y))
```
