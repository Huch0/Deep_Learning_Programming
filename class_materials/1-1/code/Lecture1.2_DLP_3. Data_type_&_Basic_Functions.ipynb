```python
import torch
```

**1.1 PyTorch의 데이터 타입**

```python
# 정수형 데이터 타입 코드 표현 실습
# 8비트 부호가 없는 정수 코드 표현 실습

a = torch.tensor(1, dtype = torch.uint8)
print('a = ', a)
print('a.dtype =', a.dtype)
```

```python
# 8비트 부호 있는 정수 코드 표현 실습

b = torch.tensor(-1, dtype = torch.int8)
print('b = ', b)
print('b.dtype =', b.dtype)
```

```python
# 8비트 부호 없는 정수 코드 표현 실습

c = torch.tensor(-1, dtype = torch.uint8)
print('c = ', c)
print('c.dtype =', c.dtype)
```

```python
# 16비트 부호 있는 정수 코드 표현 실습

d = torch.tensor(1, dtype = torch.short)
print('d = ', d)
print('d.dtype =', d.dtype)
```

```python
# 32비트 부호 있는 정수 코드 표현 실습

e = torch.tensor(2, dtype = torch.int)
print('e = ', e)
print('e.dtype =', e.dtype)
```

```python
# 64비트 부호 있는 정수 코드 표현 실습

f = torch.tensor([[1, 2, 3],
                  [4, 5, 6]], dtype = torch.long)
print('f = ', f)
print('f.dtype =', f.dtype)
```

```python
# 실수형 데이터 타입 코드 표현 실습
# 32비트 부동 소수점 수의 코드 표현 실습

g = torch.tensor(1, dtype = torch.float32)
print('g = ', g)
print('g.dtype =', g.dtype)
```

```python
# 64비트 부동 소수점 수의 코드 표현 실습

h = torch.tensor([[1, 2, 3],
                  [4, 5, 6]], dtype = torch.double)
print('h = ', h)
print('h.dtype =', h.dtype)
```

**1.2 타입 캐스팅**

```python
# 타입 캐스팅 코드 표현 실습
# 32비트 부동 소수점 수로 변환 코드 표현 실습

i = torch.tensor([2, 3, 4], dtype = torch.int8)
j = i.float()
print('j.dtype = ', j.dtype)
```

```python
# 64비트 부동 소수점 수로 변환 코드 표현

k = i.double()
print('k.dtype = ', k.dtype)
```

**2.1 Tensor의 요소를 반환하거나 계산하는 함수**

```python
# Tensor의 요소를 반환하거나 계산하는 함수의 코드 표현 실습
# min() 함수의 코드 표현 실습

l = torch.tensor([[1, 2, 3],
                  [3, 4, 5]], dtype = torch.float)

print('torch.min(l) =', torch.min(l))
```

```python
# max() 함수의 코드 표현 실습

print('torch.max(l) =', torch.max(l))
```

```python
# sum() 함수의 코드 표현 실습

print('torch.sum(l) =', torch.sum(l))
```

```python
# prod() 함수의 코드 표현 실습

print('torch.prod(l) =', torch.prod(l))
```

```python
# mean() 함수의 코드 표현 실습

print('torch.mean(l) =', torch.mean(l))
```

```python
# var() 함수의 코드 표현 실습

print('torch.var(l) =', torch.var(l))
```

```python
# std() 함수의 코드 표현 실습

print('torch.std(l) =', torch.std(l))
```

**2.2 Tensor의 특성을 확인하는 메서드**

```python
# Tensor의 차원의 수를 확인하는 코드 표현 실습

print('l.dim() = ', l.dim())
```

```python
# Tensor의 모양(크기)를 확인하는 코드 표현 실습

l.size()
print('l.size() = ', l.size())
```

```python
l.shape
print('l.shape = ', l.shape)
```

```python
# Tensor의 요소의 총 개수를 확인하는 코드 표현 실습

l.numel()
print('l.numel() = ', l.numel())
```
