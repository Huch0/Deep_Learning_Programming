```python
import torch
```

```python
# 0-D Tensor(=Scalar)의 코드 표현 실습

a = torch.tensor(36.5)
print('a =', a)
```

```python
# 1-D Tensor(=Vector)의 코드 표현 실습

b = torch.tensor([175, 60, 81, 0.8, 0.9])
print('b = ', b)
```

```python
# 2-D Tensor(=Matrix)의 코드 표현 실습

c = torch.tensor([[77, 114, 140, 191],
                  [39, 56, 46, 119],
                  [61, 29, 20, 33]])
print('c = ', c)
```

```python
# 그레이스케일 이미지의 코드 표현 실습

import matplotlib.pyplot as plt

plt.xticks([]), plt.yticks([])
plt.imshow(c, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
# 3-D Tensor의 코드 표현 실습
# 채널별로 Tensor 생성

red_channel = torch.tensor([[255, 0],
                            [0, 255]])
green_channel = torch.tensor([[0, 255],
                              [0, 255]])
blue_channel = torch.tensor([[0, 0],
                             [255, 0]])

# 세 채널을 결합해서 하나의 3차원 텐서로 합침, (4강) Manipulation of Tensors 에서 학습

d = torch.stack((red_channel, green_channel, blue_channel), dim=2)

print('d = ', d)
```

```python
# 컬러 이미지의 코드 표현 실습

plt.xticks([]), plt.yticks([])
plt.imshow(d)
plt.show()
```
