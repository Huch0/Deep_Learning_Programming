**1.1 특정한 값으로 초기화된 Tensor 생성**

```python
import torch
```

```python
# 0으로 초기화된 Tensor를 생성하는 코드 표현 실습

a = torch.zeros(5)

print('a =', a)
```

```python
print('a.size() =', a.size())
```

```python
print('a.dtype =', a.dtype)
```

```python
b = torch.zeros([2, 3])

print('b =', b)
```

```python
print('b.size() =', b.size())
```

```python
print('b.dtype =', b.dtype)
```

```python
c = torch.zeros([3, 2, 4])

print('c = ', c)
```

```python
# 1로 초기화된 Tensor를 생성하는 코드 표현 실습

d = torch.ones(3)

print('d =', d)
```

```python
e = torch.ones([3, 2])

print('e =', e)
```

```python
f = torch.ones([3, 2, 3])

print('f =', f)
```

**1.2 특정한 값으로 초기화된 Tensor 변환**

```python
# 크기와 자료형이 같은 0으로 초기화된 Tensor로 변환하는 코드 표현 실습

g = torch.zeros_like(e)

print('g =', g)
```

```python
# 크기와 자료형이 같은 1로 초기화된 Tensor로 변환하는 코드 표현 실습

h = torch.ones_like(b)

print('h =', h)
```

**1.3 난수로 초기화된 Tensor 생성**

```python
# [0, 1] 구간의 연속균등분포 난수 Tensor 생성 코드 표현 실습

i = torch.rand(3)
print('i = ', i)
```

```python
j = torch.rand([2, 3])
print('j = ', j)
```

```python
# 표준정규분포 난수 Tensor 생성 코드 표현 실습

k = torch.randn(3)
print('k = ', k)
```

```python
l = torch.randn([2, 3])
print('l = ', l)
```

```python
# 크기와 자료형이 같은 [0, 1] 구간의 연속균등분포 난수 Tensor로 변환하는 코드 표현 실습

m = torch.rand_like(k)
print('m = ', m)
```

```python
# 크기와 자료형이 같은 [0, 1] 구간의 표준정규분포 Tensor로 변환하는 코드 표현 실습

n = torch.randn_like(i)
print('n = ', n)
```

**1.4 지정된 범위 내에서 초기화된 Tensor 생성**

```python
# 지정된 범위 내에서 초기화된 Tensor 생성하는 코드 표현 실습

o = torch.arange(start = 1, end = 11, step = 2)
print('o =', o)
```

```python
o = torch.arange(1, 11, 2)
print('o =', o)
```

```python
o.dtype
print('o.dtype =', o.dtype)
```

```python
p = torch.arange(start = 1, end = 4, step = 0.5)
print('p = ', p)
```

```python
p = torch.arange(1, 4, 0.5)
print('p = ', p)
```

```python
p.dtype
print('p.dtype = ', p.dtype)
```

**1.5 초기화되지 않은 Tensor 생성**

```python
# 초기화 되지 않은 Tensor를 생성하는 코드 표현 실습

q = torch.empty(5)
print('q = ', q)
```

```python
torch.fill_(q, 3.0)
print('q = ', q)
```

```python
r = torch.empty([2, 3])
print('r = ', r)
```

```python
torch.fill_(r, 3.0)
print('r = ', r)
```

**1.6 list, Numpy 데이터로부터 Tensor 생성**

```python
# list 데이터로부터 Tensor를 생성하는 코드 표현 실습

s = [1, 2, 3, 4, 5, 6]

t = torch.tensor(s)

print('t =', t)
```

```python
# Numpy 데이터로부터 Tensor를 생성하는 코드 표현 실습

import numpy as np

u = np.array([[0, 1],
              [2, 3]])

v = torch.from_numpy(u)

print('v =', v)
```

```python
v = torch.from_numpy(u).float()

print('v = ', v)
```

**1.7 CPU Tensor 생성**

```python
# 정수형 CPU Tensor를 생성하는 코드 표현 실습

w = torch.IntTensor([1, 2, 3, 4, 5])
print('w =', w)
```

```python
A, B, C, D, E = 1, 2, 3, 4, 5
w = torch.IntTensor([A, B, C, D, E])
print('w =', w)
```

```python
# 실수형 CPU Tensor를 생성하는 코드 표현 실습

A, B, C, D, E = 1, 2, 3, 4, 5

x = torch.FloatTensor([A, B ,C, D, E])

print('x =', x)
```

**1.8 Tensor 복제**

```python
x = torch.tensor([1, 2, 3, 4, 5, 6])
print('x =', x)
```

```python
# Tensor를 복제하는 코드 표현 실습

y = x.clone()

print('y =', y)
```

```python
# 또 다른 Tensor를 복제하는 코드 표현 실습

z = x.detach()
print('z =', z)
```

**1.9 CUDA Tensor 생성과 변환**

```python
# CUDA Tesnor를 생성 & 변환하는 코드 표현 실습
# tensor가 현재 어떤 디바이스에 있는지 확인

a = torch.tensor([1, 2, 3])
a.device
```

```python
# CUDA를 사용할 수 있는 환경인지 확인

torch.cuda.is_available()
```

```python
# CUDA 디바이스 이름 확인

torch.cuda.get_device_name(device=0)
```

```python
# 사용 가능한 GPU 개수 확인

torch.cuda.device_count()
```

```python
# Tensor를 GPU에 할당

b = torch.tensor([1, 2, 3, 4, 5]).to(device = 'cuda')
print('b_gpu = ', b)
```

```python
b = torch.tensor([1, 2, 3, 4, 5]).cuda()
print('b_gpu = ', b)
```

```python
# GPU에 할당된 Tensor를 CPU Tensor로 변환

c = b.to(device = 'cpu')
print('c_cpu = ', c)
```

```python
c = b.cpu()
print('c_cpu = ', c)
```
